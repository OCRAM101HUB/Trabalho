#  **Introdução**
  Esse trabalho possui o intuito de demonstrar visualmente a mudança na presença de torcedores da Nba,nos estádios de seus respectivos times, nos anos de 2019 a 2022.Com ênfase na análise dos anos em que se decorreu a pandemia do Covid-19 e seu impacto na presença de torcedores nos jogos.

```python
import pandas as pd
import statistics
import matplotlib.pyplot as plt
import numpy as np
from numpy import *
import plotly.express as px
```

```python
#Foram consideradas a atendência de torcedores nos jogos em "casa" dos times apresentados.
planilha = pd.read_excel("/content/Pasta de trabalho 14.xlsx")
display (planilha)
```

```python
planilha = pd.read_excel("/content/Pasta de trabalho 15 (1).xlsx")
display (planilha)
```

```python
#Análise gráfica do Total de Torcedores que atenderam aos jogos dos times: 76ers,Mavericks e Miami Heat, nos anos de 2019 a 2022.
grafico = px.line(planilha, x="Anos", y="76ers")
grafico.show ()
grafico = px.line(planilha, x="Anos", y="Mavericks")
grafico.show ()
grafico = px.line(planilha, x="Anos", y="Heat")
grafico.show ()

```

  Como podemos ver nos gráficos,o número de torcedores atendendo aos jogos diminuiu nos anos de 2020 e 2021,sendo estes os anos em que estavam ocorrendo a pandemia do Covid-19.O início da temporada regular de 2020–21 foi adiado por causa da pandemia.O comissário da NBA, Adam Silver, sugeriu adiar ainda mais a temporada até pelo menos janeiro porque as ordens de saúde locais em cada cidade da NBA limitariam a participação dos fãs.

```python
planilha = pd.read_excel("/content/Pasta de trabalho 16 (1).xlsx")
display (planilha)
```

```python
#Gráfico para mostrar qual time dentre os selecionados  teve uma maior recuperação de fãs pós pandemia.
#Os times selecionados são os que costumam liderar o ranking em número de torcedores presentes nos jogos.
fig=plt.figure(figsize=(15, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(planilha.Ano, planilha.Bulls, label = 'Bulls')
plt.plot(planilha.Ano, planilha.Mavericks, label= 'Mavericks')
plt.plot(planilha.Ano, planilha.Raptors, label= 'Raptors')
plt.plot(planilha.Ano, planilha.Heat, label='Heat')
plt.plot(planilha.Ano, planilha.Warriors, label='Warriors')
plt.plot(planilha.Ano, planilha.Sixers, label='Sixers')

plt.legend()

```

Com a análise pode-se observar que o Chicago Bull foi o time que teve a maior presença de fãs nos jogos ao fim da temporada de 2022,seguido pelo Sixers que vinha liderando juntamente do Dallas Mavericks.O Miami Heat também teve uma grande recuperação na presença dos fãs nos estádios,ressaltando a importãncia disso para os times,uma vez que 40% da receita da Nba é através do público nas arenas. 

# **Conclusão**
Conforme o que foi apresentado, pode-se cocluir que a pandemia afetou drasticamente a presença de fãs nos estádios, sobretudo nos anos de 2020 e 2021, voltando a normalidade somente em 2022.Pode-se observar também que alguns times aumentaram seu número de fãs (ex:Bulls e Sixers) enquanto outros como o Toronto Raptors, ainda não recuperaram os números de torcedores presentes nos jogos que possuíam antes da pandemia.
